# Olympiacos 2023–2024: Player Analysis (Conference League)

## Objective

The goal of this analysis is to evaluate player contributions during Olympiacos' 2023–2024 Conference League-winning campaign.

We focus on identifying key performers, understanding goal contribution, and examining how individual players impacted the team’s success in European competition.

## Squad Definition

The player dataset is based on the UEFA List A registered for the knockout phase (January 2024).

This ensures that the analysis focuses on the players available during the most important stage of the Conference League campaign.

```python
import pandas as pd
import matplotlib.pyplot as plt

players = pd.read_csv("../data/raw/olympiacos_players_2023_2024.csv")

players.head()
```

```python
players["goals_per_90"] = players["goals"] / (players["minutes"] / 90)
players["assists_per_90"] = players["assists"] / (players["minutes"] / 90)

players["contribution"] = players["goals"] + players["assists"]
players["contribution_per_90"] = players["contribution"] / (players["minutes"] / 90)
```

```python
players.columns
```

```python
players.sort_values("contribution", ascending=False)
```

```python
players.sort_values("contribution_per_90", ascending=False)
```

```python
top = players.sort_values("contribution", ascending=False).head(8)

plt.figure()
plt.bar(top["player"], top["contribution"])

plt.title("Top Player Contributions (Goals + Assists)")
plt.xticks(rotation=45)

plt.show()
```
